```python
def get_data_features(df, col, day_realise = False): 
    df[col] = pd.to_datetime(df[col])
    df['year'] = df[col].dt.isocalendar().year.astype("int64")
    df['month'] = df[col].dt.month.astype("int64")
    df['week_num'] = df[col].dt.isocalendar().week.astype('int64')   
    
    if day_realise: 
        df['day'] = df[col].dt.day.astype('int64')
        
    df[col] = df[col].astype('str')
    return df
```

```python
import datetime
import calendar

def getDateFromWeekNum(y,w):
    d = "{}-W{}".format(y,w)
    r = dt.datetime.strptime(d + '-1', "%G-W%V-%u") #%G-W%V-%u #%Y-W%W-%w
    return pd.to_datetime(r)
```

```python
def calcRevenue(x):
    if x['sold_cnt_month'] >= x['plan']:
        res = x['gain_sum_month'] + x['back_bonus']
    else:
        res = x['gain_sum_month']
    return res

transactions_df['gainpbonus_month'] = \
transactions_df[['sold_cnt_month', 'plan', 'gain_sum_month', 'back_bonus']].apply(lambda x: calcRevenue(x), axis=1)
```

## Use Whole Transactions Dataset: pred +1 day

```python
def autoreg_retrain(df, curr_date):

    autor_sku_pred = pd.DataFrame() 

    sku_arr = df.SKU.unique().tolist()

    for i_sku in tqdm_notebook(sku_arr, total=len(sku_arr)): 

        pred_df = pd.DataFrame()
        _train_df = df.query("SKU == @i_sku")[['dates', 'price']]\
                                   .drop_duplicates().sort_values('dates').reset_index()
        _train_df.dropna(subset = ["price"], inplace = True)
        if _train_df.shape[0] >= 100:
            _lag = 20
        else:
            _lag = 2
        AR_ = AutoReg(_train_df.price, lags=_lag)
        ARmodel = AR_.fit()
        
        _size = 1
        next_dt = curr_date
#         next_dt = [_train_df['dates'].max() + dt.timedelta(days=i) for i in range(1,_size+1)]

        pred_df['dates'] = [next_dt]
        pred_df['dates'] = pred_df['dates'].astype(str)
        pred_df['SKU']   = [i_sku]*_size

        pred_df['arima_preds'] = ARmodel.predict(start = len(_train_df)+1, 
                                                 end   = len(_train_df) + _size,
                                                 dynamic=False
                                                 ).values
        
        pred_df['dates'] = pred_df['dates'].astype(str)
        
        autor_sku_pred = pd.concat((autor_sku_pred, pred_df), axis=0)
        
    return autor_sku_pred
```

## Run with Visualization

```python
def iterate_model(model, df, context_columns, strategies_lst, 
                  records=None, plot_frequency=250, avg_length=150):
    
    """Goes through online learning simulation with model."""
    #these will track values for plotting
    if records is None:
        records = dict()
        records['timesteps'] = []
        records['c_reward'] = []
        records['cumulative_reward'] = 0
        records['m_reward'] = []
        records['maximum_reward'] = 0
        records['regret_record'] = []
        records['avg_regret'] = []
        start = 0
    else:
        start = records['timesteps'][-1]
    for i, row in tqdm_notebook(df.iterrows(), total=df.shape[0]):
        records['timesteps'].append(i)
        
        #prepare features for model
        context = row[context_columns].values
        best_choice = strategies_lst[0]
        
        #get reward for 'best' choice
        mx = row['price']*(1.+strategies_lst[row['action']])
        records['maximum_reward'] += mx
        records['m_reward'].append(records['maximum_reward'])
        
        # make and action
        action = model.action(context)
        
        #get reward for the action chosen by model
        reward = row['price']*(1.+strategies_lst[action])
        
        #regret is the opportunity cost of not choosing the optimal promotion
        regret = mx - reward
        records['regret_record'].append(regret)
        records['cumulative_reward'] += reward
        records['c_reward'].append(records['cumulative_reward'])
        model.update(context, action, reward)
        
        #plot occasionally
        if i <= avg_length:
            if i < avg_length:
                moving_avg=0
            else:
                moving_avg = np.array(records['regret_record']).mean()
            if i == avg_length:
                records['avg_regret'] = [moving_avg] * avg_length
        else:
            moving_avg = sum(records['regret_record'][-avg_length:])/avg_length
        records['avg_regret'].append(moving_avg)
        if i % plot_frequency == 0 and i > 0:
            c_rewardplt = np.array(records['c_reward'])/max(records['m_reward'])
            m_rewardplt = np.array(records['m_reward'])/max(records['m_reward'])
            regretplt = np.array(records['avg_regret'])/max(records['avg_regret'])
            plt.plot(records['timesteps'], c_rewardplt, label='cumulative reward')
            plt.plot(records['timesteps'], m_rewardplt, label='maximum reward')
            plt.plot(records['timesteps'], regretplt, color='red', label='mean regret')
            plt.title('Normalized Reward & Regret')
            plt.legend()
            plt.show()
    return records
            
    
records = iterate_model(model=linear_model, 
                        df=transactions_df, 
                        context_columns=context_columns, 
                        strategies_lst=strategies_lst, 
                        avg_length=1000, 
                        plot_frequency=100000)
```
