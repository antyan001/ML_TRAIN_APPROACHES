```python
def get_data_features(df, col, day_realiaze = False): 
    df[col] = pd.to_datetime(df[col])
    df['year'] = df[col].dt.isocalendar().year.astype("int64")
    df['month'] = df[col].dt.month.astype("int64")
    df['week_num'] = df[col].dt.isocalendar().week.astype('int64')   
    
    if day_realiaze: 
        df['day'] = df[col].dt.day.astype('int64')
        
    df[col] = df[col].astype('str')
    return df
```

```python
%%time
from sklearn.model_selection import TimeSeriesSplit, ShuffleSplit
cv = TimeSeriesSplit(n_splits=5)
#cv = ShuffleSplit(n_splits=5, test_size=0.20)

params = {
            'min_child_weight': np.random.uniform(0.1, 15, 20),
            'gamma': np.random.uniform(0.1, 5, 10),
            'subsample': np.random.uniform(0.4, 1, 10),
            'colsample_bytree': [0.6, 0.8, 1.0],
            'max_depth': np.arange(3, 15)
        }

model = xgb.XGBRegressor(n_jobs=1, n_estimators=100, seed=17, cat_features = cat_features)

random_search = RandomizedSearchCV(model,
                                   param_distributions=params,
                                   n_iter=10,
                                   scoring='neg_mean_squared_error',
                                   n_jobs=1,
                                   cv=cv,
                                   verbose=3,
                                   random_state=17)

random_search.fit(x_train, y_train)

print(f"Best Score: {random_search.best_score_}")
print(f"Best Parames: {random_search.best_params_}")
```
