## Imputing NAN

```python
def makeImputing(df=None, strategy='constant', fill_value=-99, all_cols=None):
    
    im = SimpleImputer(strategy=strategy, fill_value=fill_value, copy=False)
    
    missing_df = df.isnull().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df = missing_df.loc[(missing_df['missing_count'] > 0), :]
    missing_df = missing_df.sort_values(by='missing_count')

    res = im.fit_transform(df[missing_df.column_name.values.tolist()])
    im_out_df = pd.DataFrame(res, columns=missing_df.column_name.values.tolist())  
    
    part_cols = [col for col in all_cols if col not in (im_out_df.columns.values)]
    out = pd.concat([df[part_cols],im_out_df], axis=1) 
    
    return out[all_cols]
```

## Preproc

```python
def makeFE(df):
    
    df = df.copy()

#     df['year_month'] = df['timestamp'].apply(lambda x: 
#                                                      dt.datetime.strftime(dt.datetime.strptime(x, '%Y-%m-%d'), 
#                                                                           format='%Y-%m'
#                                                                          )
#                                             ).astype(str)

    df['timestamp'] = pd.to_datetime(df['timestamp'])    

    # get year
    df['year'] = df.timestamp.dt.isocalendar().year.astype(int)  

    # get month of year
    df['month'] = df.timestamp.dt.month.astype(int)  

    # get day of week
    df['week_of_year'] = df.timestamp.dt.isocalendar().week.astype(int)   

    # get week of the year
    df['day_of_week'] = df.timestamp.dt.weekday.astype(int)    

#     df['timestamp_int'] = df.timestamp.view(int)    
    
    df["floor_dash_max_floor"] = df["floor"] / df["max_floor"]
    df["floor_diff_max_floor"] = df["floor"] - df["max_floor"]
    
    df["green_dash_prom"] = df["green_part"] / df["prom_part"]
    df["green_dash_indust"] = df["green_part"] / (df["indust_part"]*100.)
    
    df["hospital_beds_dash_population"] = df["hospital_beds_raion"] / df["population"]
    
    df["max_floor_dash_population"] = df["max_floor"] / df["population"]
    df["cafe_dash_population"] = df["cafe_count"] / df["population"]
    df["office_dash_population"] = df["office_num"] / df["population"]
    df["cafe_dash_office"] = df["cafe_count"] / df["office_num"]
    df["cafe_dash_shopping_cent"] = df["cafe_count"] / df["shopping_centers_facilities"]
    df["cafe_dash_sport_obj"] = df["cafe_count"] / df["sport_objects_facilities"]
    
    df["ratio_rooms_dash_full_sq"] = df["num_room"] / df["full_sq"]
    df["ratio_rooms_dash_life_sq"] = df["num_room"] / df["life_sq"]
    df["ratio_life_dash_full_sq"] = df["life_sq"] / df["full_sq"]
    df["ration_kitchen_dash_full_sq"] = df["kitch_sq"] / df["full_sq"]
    df["ration_kitchen_dash_life_sq"] = df["kitch_sq"] / df["life_sq"]

    # добавим воздраст здания
#     df["build_year_dt"] = df["build_year"].apply(lambda x: str(int(x))+'-01-01')
#     df["build_year_dt"] = pd.to_datetime(df["build_year_dt"], errors='coerce')
#     df["build_year_dt"] = df["build_year_dt"].fillna(pd.to_datetime('1900-01-01'))

    df['age'] = df["build_year"] - df['year']
    
#     df['build_year_dt'] = df["build_year_dt"].apply(lambda x: dt.datetime.timestamp(x))
#     df['timestamp_dt'] = df["timestamp"].apply(lambda x: dt.datetime.timestamp(x))
    
#     df['diff_build_ts_days'] = ((df['timestamp_dt'] - df['build_year_dt']) / (24 * 60 * 60)).astype(int)

    # добавим разность между общей и жилой площадью квартиры
    df['full_sqr_diff_life_sq'] = df["full_sq"] - df["life_sq"]
    df['full_sqr_diff_kitch_sq'] = df["full_sq"] - df["kitch_sq"]
    
    df["green_part_diff_indust"] = df["green_part"] - (df["indust_part"]*100.)
    df["green_part_diff_prom"]   = df["green_part"] - df["prom_part"]
    df['health_plus_leisure']    = df["healthcare_facilities"] + df["leisure_facilities"]
    
    df.drop(['timestamp'], axis =1, inplace=True)
    
    return df
```

```python
def makeScale(df, all_cols, exclude_cols):
    
    _df = df.copy()
    dtypes_dct = dict(df.dtypes.to_frame('dtypes').reset_index().values)
    
    numerical_ix = _df.drop(exclude_cols, axis=1).select_dtypes(include=['int64', 'float64']).columns.values.tolist()       
    cat_ix = _df.drop(exclude_cols, axis=1).select_dtypes(include=['object']).columns.values.tolist()       
      
    t = [('num', MinMaxScaler(), numerical_ix), ('cat', OrdinalEncoder(), cat_ix)]
    col_transform = ColumnTransformer(transformers=t, remainder='passthrough')
    encoder = col_transform.fit(_df)
    res = encoder.transform(_df)          
    part_cols = [col for col in all_cols if col not in numerical_ix+cat_ix]   
    reorder_df = pd.DataFrame(res, columns=numerical_ix+cat_ix+part_cols).astype(dtypes_dct)
    
    reorder_df[cat_ix] = reorder_df[cat_ix].astype(float) 
    
    return reorder_df[all_cols]
```

## Feature Binarizer

```python
def feat_binarizer(df):
    
    #full_sq
    bins = [0, 30, 40, 50, 60, 70, 80, 90, 100, 200, 5326]
    df['full_sq_bins'] = np.searchsorted(bins, df.full_sq.values)

    #life_sq
    bins = [0, 30, 40, 50, 60, 70, 80, 90, 100, 200, 800]
    df['life_sq_bins'] = np.searchsorted(bins, df.life_sq.values)

    #kitch_sq
    bins = [0, 5, 10, 20, 30, 100, 200, 2000]
    df['kitch_sq_bins'] = np.searchsorted(bins, df.kitch_sq.values)

    #cafe_count
    bins = [0, 5, 10, 20, 30, 100]
    df['cafe_count_bins'] = np.searchsorted(bins, df.cafe_count.values)

    #office_num
    bins = [0, 5, 10, 20, 30, 100]
    df['office_num_bins'] = np.searchsorted(bins, df.office_num.values)
    
    #floor
    bins = [0, 5, 10, 20, 30, 50, 70]
    df['floor_bins'] = np.searchsorted(bins, df.floor.values)    

    #max_floor
    bins = [0, 5, 10, 20, 30, 50, 70, 100]
    df['max_floor_bins'] = np.searchsorted(bins, df.max_floor.values)    
    
    df.drop(['full_sq', 
             'life_sq', 
             'kitch_sq', 
             'cafe_count', 
             'office_num',
             'floor',
             'max_floor'], axis=1, inplace=True)
    
    return df
```

## Work With Outliers in Train

```python
def cap_data(df, excld_col):
    for col in df.drop([excld_col], axis=1).columns:
        print("capping the ",col)
        if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
            percentiles = df[col].quantile([0.01,0.99]).values
            df.loc[df[col] <= percentiles[0],col] = percentiles[0]
            df.loc[df[col] >= percentiles[1],col] = percentiles[1]
        else:
            df.loc[:,col]=df.loc[:,col]
    return df

def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

def fill_outliers(df):
    
    df_all = df.copy()
    numerical_ix = df_all.select_dtypes(include=['int64', 'float64']).columns.values.tolist() 
    
    for col in numerical_ix:
        # Detect outliers in this column
        data = df_all[col].values.copy()
        data_mean, data_std = np.mean(data), np.std(data, ddof=0)
        print('{} --> mean: {} std: {}'.format(col, data_mean, data_std))
        cut_off = data_std * 3
        lower, upper = data_mean - cut_off, data_mean + cut_off
        mask_outliers = [True if (x <= lower) or (x >= upper) else False for x in data]
        df_all.loc[mask_outliers, col] = -99
    
    return df_all
```
