```python
train_main_df['YYdd'] = train_main_df['timestamp'].apply(lambda x: 
                                                         dt.datetime.strftime(dt.datetime.strptime(x, '%Y-%m-%d'), 
                                                                              format='%Y-%m'
                                                                             )
                                                        )
```

```python
out = train_main_df.groupby(['YYdd']).agg({'price':'mean'}).sort_values(by='YYdd',ascending=True).reset_index()
out.head()
```

```python
g=sns.pairplot(x_vars='YYdd', y_vars='price', data=out)
g.fig.set_figheight(8)
g.fig.set_figwidth(14)
plt.ylabel('mean_price', fontsize=12)
plt.xlabel('YY-dd', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Посчитать количество пропусков в life_sq.  
train_main_df['life_sq'].isnull().sum(axis=0)
```

```python
from sklearn.preprocessing import StandardScaler, MaxAbsScaler
from sklearn.impute import SimpleImputer
from scipy.stats.stats import pearsonr 
```

```python
missing_df = train_main_df.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df = missing_df.loc[(missing_df['missing_count'] > 0), :]
missing_df = missing_df.sort_values(by='missing_count')
```

```python
im = SimpleImputer(strategy='mean', fill_value=None, copy=False)
```

```python
res = im.fit_transform(train_main_df[missing_df.column_name.values.tolist()])
im_out_df = pd.DataFrame(res, columns=missing_df.column_name.values.tolist())

im_out_df.head(10)
```

```python
#Сохранить коэффициенты корреляции Пирсона между (price, full_sq) и (price, life_sq без пропусков).
```

```python
pearsonr(train_main_df.price.values, train_main_df.full_sq.values)[0]
```

```python
pearsonr(train_main_df.price.values, im_out_df.life_sq.values)[0]
```
