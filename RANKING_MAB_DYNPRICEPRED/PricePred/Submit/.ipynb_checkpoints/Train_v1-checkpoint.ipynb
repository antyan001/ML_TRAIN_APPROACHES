## Imputing NAN

```python
def makeImputing(df=None, strategy='mean', all_cols=None):
    
    im = SimpleImputer(strategy=strategy, fill_value=None, copy=False)
    
    missing_df = df.isnull().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df = missing_df.loc[(missing_df['missing_count'] > 0), :]
    missing_df = missing_df.sort_values(by='missing_count')

    res = im.fit_transform(df[missing_df.column_name.values.tolist()])
    im_out_df = pd.DataFrame(res, columns=missing_df.column_name.values.tolist())  
    
    part_cols = [col for col in all_cols if col not in (im_out_df.columns.values)]
    out = pd.concat([df[part_cols],im_out_df], axis=1) 
    
    return out[all_cols]
```

```python
imp_whole_df      = makeImputing(whole_train_df, 'mean', all_cols=all_cols_tr)
imp_whole_df_test = makeImputing(whole_test_df, 'mean', all_cols=all_cols_test)
```

## Preproc

```python
def makeFE(df):
    
    df = df.copy()

    df['year_month'] = df['timestamp'].apply(lambda x: 
                                                     dt.datetime.strftime(dt.datetime.strptime(x, '%Y-%m-%d'), 
                                                                          format='%Y-%m'
                                                                         )
                                            ).astype(str)

    df['timestamp'] = pd.to_datetime(df['timestamp'])    

    # get year
    df['year'] = df.timestamp.dt.isocalendar().year.astype(int)  

    # get month of year
    df['month'] = df.timestamp.dt.month.astype(int)  

    # get day of week
    df['week_of_year'] = df.timestamp.dt.isocalendar().week.astype(int)   

    # get week of the year
    df['day_of_week'] = df.timestamp.dt.weekday.astype(int)    

    df['timestamp_int'] = df.timestamp.view(int)    
   
    df["ratio_life_dash_full_sq"] = df["life_sq"] / df["full_sq"]
    df["ration_kitchen_dash_full_sq"] = df["kitch_sq"] / df["full_sq"]

    # добавим воздраст здания
    df['age'] = df["build_year"] - df['year']

    # добавим разность между общей и жилой площадью квартиры
    df['some_extra_sqr'] = df["full_sq"] - df["life_sq"]    

    return df
```

```python
def makeScale(df, all_cols, exclude_cols):
    
    _df = df.copy()
    dtypes_dct = dict(df.dtypes.to_frame('dtypes').reset_index().values)
    
    numerical_ix = _df.drop(exclude_cols, axis=1).select_dtypes(include=['int64', 'float64']).columns.values.tolist()       
    t = [('num', MinMaxScaler(), numerical_ix)]
    col_transform = ColumnTransformer(transformers=t, remainder='passthrough')
    encoder = col_transform.fit(_df)
    res = encoder.transform(_df)          
    part_cols = [col for col in all_cols if col not in numerical_ix]   
    reorder_df = pd.DataFrame(res, columns=numerical_ix+part_cols).astype(dtypes_dct)
    
    return reorder_df[all_cols]
```

## Train XGBOOST

```python
fin_scl_tr.timestamp.describe(datetime_is_numeric=True)
```

```python
train_y = fin_scl_tr[fin_scl_tr.timestamp < pd.to_datetime('2015-05-01')].price
val_y = fin_scl_tr[fin_scl_tr.timestamp >= pd.to_datetime('2015-05-01')].price

train_X = fin_scl_tr[fin_scl_tr.timestamp < pd.to_datetime('2015-05-01')].drop(['id','price','timestamp', 'year_month'], axis=1)
val_X   = fin_scl_tr[fin_scl_tr.timestamp >= pd.to_datetime('2015-05-01')].drop(['id','price','timestamp', 'year_month'], axis=1)
```

```python
train_X.shape, val_X.shape
```

```python
col_list = train_X.columns.values
col_list
```

```python
xgb_train = xgb.DMatrix(train_X[col_list], train_y, feature_names = col_list, enable_categorical=True)
xgb_test  = xgb.DMatrix(val_X[col_list], val_y, feature_names = col_list, enable_categorical=True)

evallist = [(xgb_test, 'eval'), (xgb_train, 'train')]
```

```python
xgb_params = {
    'eta': 0.05,
    'max_depth': 4,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'min_child_weight':1,
    'silent': 1,
    'seed':0
}
```

```python
model = xgb.train(params = xgb_params,
                  dtrain = xgb_train, 
                  num_boost_round = 5000, 
                  evals = evallist, 
                  early_stopping_rounds = 20, 
                  verbose_eval = 10)
```

## GidSearch

```python
X_train, X_test, y_train, y_test = train_test_split(
    fin_tr.drop(['id','price','timestamp', 'year_month'], axis=1),
    fin_tr.price, test_size=1425, random_state=42)
```

```python
cv = 2

paramGrid = \
{
    'eta': [0.5, 0.2, 0.1, 0.05, 0.01],
    'max_depth': [20, 15, 10, 8, 5],
    'subsample': [0.9, 0.8, 0.7, 0.6, 0.5],
    'colsample_bytree': [0.9, 0.8, 0.7, 0.6, 0.5],
    'min_child_weight':[0.9, 0.8, 0.7, 0.6, 0.5],
    'objective': ['reg:squarederror']*5
}

fit_params={"early_stopping_rounds":20, 
            "eval_metric" : "rmse", 
            "eval_set" : [[X_test, y_test]]}

model = xgb.XGBRegressor()

#StratifiedKFold(n_splits=cv, shuffle=True).get_n_splits([X_train, y_train])
gridsearch = GridSearchCV(model, paramGrid, verbwose=20,             
                          cv=TimeSeriesSplit(n_splits=cv, test_size=1200).split(X_train, y_train),
                          return_train_score=True)

gridsearch.fit(X_train, y_train, **fit_params)
```
